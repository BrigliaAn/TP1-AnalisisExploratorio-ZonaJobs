```python
print(p_genero_edad.shape)
print(p_educacion.shape)
```

```python
postulantes = pd.merge(p_educacion,p_genero_edad,on="idpostulante",how="outer")
postulantes.head()
```

```python
#cantidad de postulantes agrupado por nivel de educación
postulantes.groupby('nombre').agg('count').sort_values('idpostulante',ascending=False)
```

```python
print(postulantes['estado'].unique())
print(avisos_detalle['tipo_de_trabajo'].unique())
```

```python
#cantidad de avisos por tipo de trabajo (full time, part time...)
avisos_detalle.groupby('tipo_de_trabajo').agg({'count'})
```

```python
#avisos con mayores postulantes

grouped = postulaciones.groupby('idaviso').agg({'idpostulante':'size'}).sort_values('idpostulante',ascending=False).head(10)
grouped = grouped.reset_index()
pd.merge(grouped,avisos_detalle,on="idaviso",how="left")
```

```python
#empresas con mayor cantidad de avisos
avisos_detalle.groupby('denominacion_empresa').agg({'denominacion_empresa':'size'}).sort_values('denominacion_empresa',ascending=False).head(20)
```

```python
#distribucion de postulaciones por tiempo, mes, semana, etc
postulaciones['fechapostulacion'] = pd.to_datetime(postulaciones['fechapostulacion'])
print(postulaciones['fechapostulacion'].max())
print(postulaciones['fechapostulacion'].min())
visitas['timestamp'] = pd.to_datetime(visitas['timestamp'])
print(visitas['timestamp'].max())
print(visitas['timestamp'].min())
```

```python
postulaciones_enero = postulaciones.loc[postulaciones['fechapostulacion'].dt.month == 1]
postulaciones_febrero = postulaciones.loc[postulaciones['fechapostulacion'].dt.month == 2]
```
