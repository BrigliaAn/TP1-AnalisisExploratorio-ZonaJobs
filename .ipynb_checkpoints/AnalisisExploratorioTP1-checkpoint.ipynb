```python
print(p_genero_edad.shape)
print(p_educacion.shape)
```

```python
postulantes = pd.merge(p_educacion,p_genero_edad,on="idpostulante",how="outer")
postulantes.head()
```

```python
#cantidad de postulantes agrupado por nivel de educación
postulantes.groupby('nombre').agg('count').sort_values('idpostulante',ascending=False)
```

```python
fig = plt.figure()

ax = p_educacion['nombre'].value_counts().plot(kind='bar',figsize=(13,6));

plt.title('Cantidad de postulantes por nivel educativo alcanzado', fontsize=24)

ax.set_xlabel('Nivel educativo', fontsize= 16)
ax.set_ylabel('Cantidad de postulantes', fontsize= 16);
```

```python
print(postulantes['estado'].unique())
print(avisos_detalle['tipo_de_trabajo'].unique())
```

```python
#cantidad de avisos por tipo de trabajo (full time, part time...)
avisos_detalle.groupby('tipo_de_trabajo').agg({'count'})
```

```python
fig = plt.figure()

ax = avisos_detalle['tipo_de_trabajo'].value_counts().plot(kind='bar',figsize=(13,6));

plt.title('Cantidad de avisos por tipo de trabajo', fontsize=24)

ax.set_xlabel('Tipo de trabajo', fontsize= 16)
ax.set_ylabel('Cantidad de avisos', fontsize= 16);
```

```python
#avisos con mayores postulantes

grouped = postulaciones.groupby('idaviso').agg({'idpostulante':'size'}).sort_values('idpostulante',ascending=False).head(10)
grouped = grouped.reset_index()
merge = pd.merge(grouped,avisos_detalle,on="idaviso",how="left")
merge
```

```python
merge.plot(x='titulo', y='idpostulante', kind='bar',figsize=(13,6))
#idpostulante no es el id sino la cantidad de postulantes
#en el eje x puse el tiulo del aviso pero quizas en este caso deberia ser el id si lo que se quiere saber es
#exactamente eso
```

```python
#Este es parecido al que hizo Anto solo que en lugar de ser el aviso (idaviso) con mayor postulante, este tiene en
#cuenta el titulo sin importar de que aviso de cualquier empresa venga
merge2 = pd.merge(postulaciones,avisos_detalle,on="idaviso",how="left")

fig = plt.figure()

ax = merge2['titulo'].value_counts().head(10).plot(kind='bar',figsize=(13,6));

plt.title('Trabajos con mayores postulantes', fontsize=24)

ax.set_xlabel('Trabajos', fontsize= 16)
ax.set_ylabel('Cantidad de postulantes', fontsize= 16);
```

```python
#empresas con mayor cantidad de avisos
avisos_detalle.groupby('denominacion_empresa').agg({'denominacion_empresa':'size'}).sort_values('denominacion_empresa',ascending=False).head(20)
```

```python
fig = plt.figure()

ax = avisos_detalle['denominacion_empresa'].value_counts().head(20).plot(kind='bar',figsize=(13,6));

plt.title('Top 20 de empresas con mayor cantidad de avisos', fontsize=24)

ax.set_xlabel('Empresas', fontsize= 16)
ax.set_ylabel('Cantidad de avisos', fontsize= 16);
```
